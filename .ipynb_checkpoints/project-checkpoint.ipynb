```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('satcat.tsv', sep='\t', low_memory=False)
```

```python
orgs = pd.read_csv("orgs.tsv", sep="\t", usecols=["#Code", "Class", "ShortName"]).drop(0)
orgs.rename(columns={"#Code": "Owner"}, inplace=True)
```

```python
class_mapping = {
    "A": "Academic/Non-Profit",
    "B": "Business/Commercial",
    "C": "Civil Government",
    "D": "Defense/Military"
}

orbit_mapping = {
    'LEO': ['LEO/I', 'LLEO/I', 'LEO/P', 'LLEO/P', 'LEO/S', 'LEO/E', 'LEO/R', 'LLEO/R', 'LLEO/S', 'LLEO/E'],
    'MEO': ['MEO'],
    'GEO': ['GEO/NS', 'GEO/T', 'GEO/S', 'GEO/D', 'GEO/I', 'GEO/ID'],
    'GTO': ['GTO'],
    'HEO': ['HEO', 'VHEO', 'HEO/M'],
    'Deep Space': ['DSO', 'SO', 'CLO', 'TA'],
    'Unclassified': ['-']
}

def get_class_for_owner(owner):
    owners = owner.split("/")
    classes = orgs[orgs["Owner"].isin(owners)]["Class"].unique()
    
    return classes[0] if len(classes) > 0 else None
    
def map_orbit_type(oporbit):
    for category, types in orbit_mapping.items():
        if oporbit in types:
            return category
    return 'Other'

df['OrbitType'] = df['OpOrbit'].apply(map_orbit_type)
df["Class"] = df["Owner"].apply(get_class_for_owner)
df = df[df['Status'] == 'O']
```

```python
df = df.merge(orgs[['Owner', 'ShortName']], on='Owner', how='left')
```

```python
# Count number of satellites per country
country_counts = df['State'].value_counts().reset_index()
country_counts.columns = ['Country', 'Satellite_Count']

# Approximate country coordinates (You can refine this with a geolocation dataset)
country_coords = {
    'US': (-98.35, 39.50),   # USA
    'RU': (105.32, 61.52),   # Russia
    'CN': (104.19, 35.86),   # China
    'IN': (78.96, 20.59),    # India
    'EU': (10.45, 51.17),    # Europe (approximate)
    'JP': (138.25, 36.20),   # Japan
    'UK': (-3.43, 55.38),    # United Kingdom
    'FR': (2.21, 46.23),     # France
    'DE': (10.45, 51.17),    # Germany
    'BR': (-51.93, -14.23),  # Brazil
}

# Merge country counts with coordinates
country_counts['Coordinates'] = country_counts['Country'].map(country_coords)
country_counts = country_counts.dropna()  # Remove missing locations

# Extract lat/lon
lons, lats, sizes = [], [], []
for _, row in country_counts.iterrows():
    lon, lat = row['Coordinates']
    lats.append(lat)
    lons.append(lon)
    sizes.append(row['Satellite_Count'])  # Bubble size is satellite count

# Normalize bubble sizes for better visualization
sizes = [s for s in sizes]  # Adjust multiplier for clarity

# Plot world map with bubbles
plt.figure(figsize=(12, 6))
m = Basemap(projection='robin', lon_0=0, resolution='c')

m.drawcoastlines()
m.drawcountries()
m.fillcontinents(color='lightgray', lake_color='lightblue')
m.drawmapboundary(fill_color='lightblue')

# Convert lat/lon to map coordinates
x, y = m(lons, lats)

# Plot bubbles
m.scatter(x, y, s=sizes, color='red', alpha=0.6, edgecolors='k', linewidth=0.5)

# Title and legend
plt.title("Global Satellite Distribution - Bubble Map")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.dates import YearLocator, DateFormatter

# Assuming your dataframe is called 'df'
# Filter for US satellites and sort by launch date
us_satellites = df[(df['State'] == 'US') & (df['LDate'].notna())].sort_values('LDate')

# Convert LDate to datetime
us_satellites['LDate'] = pd.to_datetime(us_satellites['LDate'])

# Group smaller categories
def group_small_categories(x):
    return x if x in ['B', 'C', 'D'] else 'A'

us_satellites['Class_grouped'] = us_satellites['Class'].apply(group_small_categories)

# Create a pivot table with cumulative counts
pivot_data = us_satellites.pivot_table(values='#JCAT', index='LDate', columns='Class_grouped', aggfunc='count', fill_value=0).cumsum()

# Calculate percentage for each class
pivot_data_percentage = pivot_data.div(pivot_data.sum(axis=1), axis=0) * 100

# Set up the plot style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Create the stacked area chart
fig, ax = plt.subplots(figsize=(15, 8))
pivot_data_percentage.plot.area(stacked=True, ax=ax, alpha=0.8)

plt.title('Relative Distribution of US Satellite Launches by Class Over Time', fontsize=16, pad=20)
plt.xlabel('Launch Date', fontsize=12)
plt.ylabel('Percentage of Satellites', fontsize=12)
plt.legend(title='Satellite Class', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)
plt.ylim(0, 100)  # Set y-axis limits to 0-100%

# Improve x-axis ticks
ax.xaxis.set_major_locator(YearLocator(5))
ax.xaxis.set_major_formatter(DateFormatter('%Y'))
plt.xticks(rotation=45)

# Remove top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Filter dataset for US satellites with a valid Class
us_satellites = df[(df['State'] == 'US') & (df['Class'].notna())]

# Count satellites per owner within each industry (Class)
owner_counts = us_satellites.groupby(['Class', 'Owner']).size().reset_index(name='count')

# Define the number of top owners per industry to display
N = 5  # Adjust this as needed

# Get the top N owners per class
top_owners = owner_counts.groupby('Class').apply(lambda x: x.nlargest(N, 'count')).reset_index(drop=True)

top_owners["scaled_count"] = top_owners.groupby("Class")["count"].transform(lambda x: x / x.max() * 100)
```

```python
def get_label_rotation(angle, offset):
    """Determine rotation and alignment for labels."""
    rotation = np.rad2deg(angle + offset)
    if angle <= np.pi:
        alignment = "right"
        rotation += 180
    else:
        alignment = "left"
    return rotation, alignment

def add_labels(angles, values, labels, offset, ax):
    """Add labels to the bars."""
    padding = 4  # Space between bar and text
    for angle, value, label in zip(angles, values, labels):
        rotation, alignment = get_label_rotation(angle, offset)
        ax.text(
            x=angle, 
            y=value, 
            s=label, 
            ha=alignment, 
            va="center", 
            rotation=rotation, 
            rotation_mode="anchor"
        )
```

```python
# Grab the group values
GROUP = top_owners["Class"].values

# Extract number of items per group
GROUPS_SIZE = [len(i[1]) for i in top_owners.groupby("Class")]

PAD = 1
ANGLES_N = len(VALUES) + PAD * len(np.unique(GROUP))
ANGLES = np.linspace(0, 2 * np.pi, num=ANGLES_N, endpoint=False)
WIDTH = (2 * np.pi) / len(ANGLES)

offset = 0
IDXS = []
for size in GROUPS_SIZE:
    IDXS += list(range(offset + PAD, offset + size + PAD))
    offset += size + PAD

VALUES = np.sqrt(top_owners["count"].values)
LABELS = top_owners["Owner"].values
# Same layout as above
fig, ax = plt.subplots(figsize=(15, 15), subplot_kw={"projection": "polar"})

ax.set_theta_offset(OFFSET)
ax.set_ylim(-100, 85)
ax.set_frame_on(False)
ax.xaxis.grid(False)
ax.yaxis.grid(False)
ax.set_xticks([])
ax.set_yticks([])

# Use different colors for each group!
GROUPS_SIZE = [len(i[1]) for i in top_owners.groupby("Class")]
COLORS = [f"C{i}" for i, size in enumerate(GROUPS_SIZE) for _ in range(size)]

# And finally add the bars. 
# Note again the `ANGLES[IDXS]` to drop some angles that leave the space between bars.
ax.bar(
    ANGLES[IDXS], VALUES, width=WIDTH, color=COLORS, 
    edgecolor="white", linewidth=2
)

# add_labels(ANGLES[IDXS], VALUES, LABELS, OFFSET, ax)

# Title
plt.title("Top Satellite Owners in Each Industry (US)", fontsize=14, fontweight='bold')

plt.show()
```

```python
from pywaffle import Waffle
```

```python
active_satellites = df[df["Status"].isin(["O", "AO", "AO IN", "OX"])]

# Count active satellites owned by the US and the rest of the world
us_count = active_satellites[active_satellites['State'] == 'US'].shape[0]
non_us_count = active_satellites[active_satellites['State'] != 'US'].shape[0]

# Define scaling factor for the waffle chart (adjustable)
TOTAL_SQUARES = 20 * 15  # 300 total squares
total_satellites = us_count + non_us_count
scale_factor = TOTAL_SQUARES / total_satellites  # Scale active satellites to waffle chart size

# Compute number of squares per category
us_squares = round(us_count * scale_factor)
non_us_squares = TOTAL_SQUARES - us_squares  # Ensure sum matches TOTAL_SQUARES
```

```python
colors = ["#4c72b0", "#dd8452"]  # US = muted blue, Non-US = muted orange

# Reverse order if US has fewer satellites
if us_squares < non_us_squares:
    values = [non_us_squares, us_squares]  # Non-US first, then US
    labels = ["Non-US Satellites", "US-Owned Satellites"]
    colors.reverse()  # Swap colors to match
else:
    values = [us_squares, non_us_squares]  # US first
    labels = ["US-Owned Satellites", "Non-US Satellites"]

# Set up figure
fig, ax = plt.subplots(figsize=(8, 6), constrained_layout=True)

# Create waffle chart
Waffle.make_waffle(
    rows=20, 
    columns=15, 
    values=values, 
    colors=colors,
    title={"label": "Active Satellites in Space: US vs. Rest of World", "loc": "left", "fontsize": 16},
    legend={
        "labels": labels,
        "loc": "center left",
        "bbox_to_anchor": (1.05, 0.5),  # Move legend outside to the right
        "fontsize": 12,
    },
    figsize=(5, 5),
    ax=ax,
    block_aspect_ratio=1,
    interval_ratio_x=0.12,  # Slightly tighter layout
    interval_ratio_y=0.12,
    vertical=True  # Ensures better alignment
)

# Show Plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data: Number of satellites by category
labels = ['Business', 'Military', 'Academic', 'Civil Government']
values = [400, 250, 150, 200]  # Example satellite counts
colors = ['#4c72b0', '#dd8452', '#55a868', '#c44e52']  # Custom colors

# Convert satellite count to dots (1 dot = 10 satellites)
dots_per_category = [v // 10 for v in values]

# Create Donut Chart
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'aspect': 'equal'})

# Define sector angles
wedges, _ = ax.pie(values, labels=labels, colors=colors, 
                    wedgeprops={'width': 0.3, 'edgecolor': 'white'}, startangle=140)

# Improved dot spacing parameters
r_min, r_max = 0.55, 0.82  # Reduce the space between rows
dot_size = 15  # Size of dots
row_spacing = 0.05  # Vertical spacing between rows
col_spacing = 0.08  # Horizontal spacing between dots

# Plot dots inside each sector
for i, wedge in enumerate(wedges):
    theta1, theta2 = np.radians(wedge.theta1), np.radians(wedge.theta2)  # Convert angles to radians

    # Determine the number of rows dynamically based on the available space
    num_rows = int(np.floor((r_max - r_min) / row_spacing))  
    num_cols = int(np.ceil(dots_per_category[i] / num_rows))  # Adjust number of columns

    # Generate row positions
    r_values = np.linspace(r_max, r_min, num_rows)  # Rows spread vertically within the sector

    # Place dots within the sector in **rows**
    for row_idx, r in enumerate(r_values):
        theta_values = np.linspace(theta1, theta2, num_cols + 2)[1:-1]  # Spread dots evenly

        for theta in theta_values:
            x, y = r * np.cos(theta), r * np.sin(theta)  # Convert polar to Cartesian coordinates
            ax.scatter(x, y, s=dot_size, color='black', alpha=0.9)  # Plot dot

# Final Adjustments
ax.set_title("Satellite Distribution by Category (1 Dot = 10 Satellites)", fontsize=14)
ax.set_xticks([]), ax.set_yticks([])  # Remove ticks
ax.set_frame_on(False)  # Hide frame

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert to datetime and filter for valid dates
df['LDate'] = pd.to_datetime(df['LDate'], errors='coerce')
df = df.dropna(subset=['LDate'])  # Remove missing dates

# Keep only US-based launches within the last 5 years
last_5_years = df[df['LDate'].dt.year >= (df['LDate'].dt.year.max() - 7)]
us_launches = last_5_years[last_5_years['State'] == 'US']

# Filter for payloads only ('Type' includes 'P')
payload_launches = us_launches[us_launches['Type'].str.contains('P', na=False) & (us_launches["Status"].isin(["O", "AO", "AO IN", "OX"]))]

# Fill missing ShortNames with Owner name (to avoid NaNs)
payload_launches['ShortName'] = payload_launches['ShortName'].fillna(payload_launches['Owner'])
payload_launches.loc[payload_launches['Owner'] == 'SPXS', 'ShortName'] = 'SpaceX'

# Count **unique** launches per company using Launch_Tag
unique_launches = payload_launches.groupby(['LDate', 'ShortName'])['Launch_Tag'].nunique().reset_index()

# Count launches per company and identify top 5
company_counts = unique_launches['ShortName'].value_counts()
top_5_companies = company_counts.nlargest(5).index.tolist()

# Group companies: Keep top 5
unique_launches = unique_launches[unique_launches['ShortName'].isin(top_5_companies)]

# Aggregate **unique launches** per year per company
launch_counts = unique_launches.groupby([unique_launches['LDate'].dt.year, 'ShortName'])['Launch_Tag'].sum().unstack(fill_value=0)

# Add a 'Total' column summing across all companies per year
launch_counts['Total'] = launch_counts.sum(axis=1)

# Define custom colors, ensuring SpaceX has a distinct, bold color
company_colors = {
    top_5_companies[0]: '#1f77b4',  # SpaceX (Bold Blue)
    top_5_companies[1]: '#ff7f0e',  # Company 2 (Orange)
    top_5_companies[2]: '#2ca02c',  # Company 3 (Green)
    top_5_companies[3]: '#d62728',  # Company 4 (Red)
    top_5_companies[4]: '#9467bd',  # Company 5 (Purple)
}

# Ensure color mapping aligns with available companies
colors = [company_colors.get(company, '#7f7f7f') for company in launch_counts.columns if company != 'Total']

# Plot stacked bar chart
ax = launch_counts.drop(columns=['Total']).plot(kind='bar', stacked=True, figsize=(6, 3), color=colors)

# Formatting
ax.set_title("US Launches by Company", fontsize=14)
ax.set_xlabel("Year")
ax.set_ylabel("Number of Unique Launches")
ax.legend(title="Company", loc="upper left", bbox_to_anchor=(1, 1))
ax.set_xticklabels(launch_counts.index, rotation=45)
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.tight_layout()
plt.show()
```
